```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#functional method
plt.plot(x,y)
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',lw=5,alpha=1,ls=':',marker='+',markersize=10,
        markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='red')

ax.set_xlim([0,2])
ax.set_ylim([0,2])
```

```python
import seaborn as sns
%matplotlib inline
tips =  sns.load_dataset('tips')
flights = sns.load_dataset( 'flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp =  flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=2)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
